```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast
import seaborn as sns

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_da_us = df[(df['job_country'] == "United States") & (df['job_title_short'] == "Data Analyst")].copy()
df_da_us["job_posted_month_no"] = df_da_us['job_posted_date'].dt.month
df_da_us_skills = df_da_us.explode('job_skills')
```

```python
df_da_us_pivot = df_da_us_skills.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_da_us_pivot.loc["Total"] = df_da_us_pivot.sum(numeric_only=True)
```

```python
df_da_us_pivot
```

```python
job_skills = df_da_us_pivot.loc["Total"].sort_values(ascending=False).index.to_list()	
```

```python
df_da_us_pivot = df_da_us_pivot[job_skills]
df_da_us_pivot = df_da_us_pivot.drop('Total', axis=0)
```

```python
df_da_us_pivot
```

```python
df_da_us_totals = df_da_us.groupby("job_posted_month_no").size() #.to_frame(name="job_count")
```

```python
df_da_us_percent = df_da_us_pivot.div(df_da_us_totals/100, axis=0)
```

```python
df_da_us_percent = df_da_us_percent.reset_index()
df_da_us_percent['job_posted_month'] = df_da_us_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_da_us_percent = df_da_us_percent.set_index('job_posted_month')
df_da_us_percent = df_da_us_percent.drop(columns='job_posted_month_no')
df_da_us_percent.iloc[:,:5]
```

```python
from adjustText import adjust_text
from matplotlib.ticker import PercentFormatter

df_plot = df_da_us_percent.iloc[:,:5]
sns.set_theme(style="ticks")
sns.lineplot(data=df_plot, dashes=False, palette="tab10", legend=False)
sns.despine()

plt.xlabel('2023')
plt.ylabel('Likelihood of Job Posting')
plt.title("Trending Top Skills for Data Analysts in the USA")

texts = []
for column in df_plot.columns:
  texts.append(plt.text(11.2, df_plot[column].iloc[-1], column))

adjust_text(texts,arrowprops=dict(arrowstyle='->', color='gray', lw=1))

plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))		

plt.show()
```
